**As we have seen there is columns which have Categorical Values. That is Geography and Gender. We have to convert it in Numeric Form. For This purpose, We have to Use label Encoder, and One Hot Encoding Functions**

### For Changing Gender Column

```python
X= churn.iloc[:,3:13].values
Y= churn.iloc[:,13].values
```

```python
pd.DataFrame(X)
```

### Label Encoder and One Hot Encoding Function

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
print(X)
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('ohe', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X), dtype = np.str) #.toarray()
print(X)
X = X[:, 1:]
print(X)
```

**The dataset is Looking Good. Now we have to split the dataset into training and test data. For this purpose,we have to use train test split function from Sklearn library**

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                    test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Now Its Time To Build Artifical Neural Network Model**

```python
from tensorflow.keras.models import Sequential 
#from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, Y_train, batch_size = 10, epochs = 300)
```

**Wow! We have got the 86% Accuracy. Its Time To Evaluate The Model.**

### Evaluate The Model

```python
classifier.evaluate(X_test, Y_test)
```

**WOW! We have got 85% Accuracy in Evaluation** 

### Confusion Matrix

```python
Y_pred = classifier.predict(X_test)
Y_pred = (Y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
accuracy_score(Y_test,Y_pred)
```

### Import Libraries

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

### Loading Sample Image and Convert it into RGB

```python
# Loading the image example and converting it to RGB as OpenCV keep the color as BGR
cat_pic_arr = cv2.imread('cat_image.jpg')
cat_pic_arr = cv2.cvtColor(cat_pic_arr, cv2.COLOR_BGR2RGB)
plt.imshow(cat_pic_arr)
```

### Using Only Red Channel From RGB Image

```python
# Using only the red chanel from the RGB image
cat_pic_red = cat_pic_arr.copy()
cat_pic_red[:, :, 1] = 0    # Zero out contribution from green
cat_pic_red[:, :, 2] = 0    # Zero out contribution from blue
plt.imshow(cat_pic_red)
```

```python
cat_pic_red[:, :, 1] 
```

### Using Only with Green Channel

```python
# Using only the green chanel from the RGB image
cat_pic_green = cat_pic_arr.copy()
cat_pic_green[:, :, 0] = 0  # Zero out contribution from red
cat_pic_green[:, :, 2] = 0    # Zero out contribution from blue
plt.imshow(cat_pic_green)
```

### Working With Blue Channel

```python
# Using only the Blue chanel from the RGB image
cat_pic_random = cat_pic_arr.copy()
cat_pic_random[:, :, 1] = 0  # Zero out contribution from red
cat_pic_random[:, :, 0] = 0    # Zero out contribution from blue
plt.imshow(cat_pic_random)
```

```python
# Using only the Blue chanel from the RGB image
cat_pic_random = cat_pic_arr.copy()
cat_pic_random[:, :, 0] = 0  # Zero out contribution from red
cat_pic_random[:, :, 0] = 0    # Zero out contribution from blue
plt.imshow(cat_pic_random)
```

```python
# Using only the Blue chanel from the RGB image
cat_pic_random = cat_pic_arr.copy()
cat_pic_random[:, :, 2] = 0  # Zero out contribution from red
cat_pic_random[:, :, 2] = 0    # Zero out contribution from blue
plt.imshow(cat_pic_random)
```

```python
# Using only the Blue chanel from the RGB image
cat_pic_random = cat_pic_arr.copy()
cat_pic_random[:, 2:, 0] = 0  # Zero out contribution from red
cat_pic_random[:, 1:, 2] = 0    # Zero out contribution from blue
plt.imshow(cat_pic_random)
```

### Displaying RGB Image

```python
## Concerting rgb values to gray
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    return  0.2989 * r  + 0.5870 * g + 0.1140 * b

gray = rgb2gray(cat_pic_arr.copy())
plt.imshow(gray, cmap='gray')
```
